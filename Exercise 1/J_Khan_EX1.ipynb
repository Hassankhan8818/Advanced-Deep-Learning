```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# Load datasets
(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = tf.keras.datasets.mnist.load_data()
(x_train_cifar, y_train_cifar), (x_test_cifar, y_test_cifar) = tf.keras.datasets.cifar10.load_data()
```

```python
# Preprocess MNIST data
x_train_mnist = x_train_mnist[..., np.newaxis] / 255.0
x_test_mnist = x_test_mnist[..., np.newaxis] / 255.0
y_train_mnist = tf.keras.utils.to_categorical(y_train_mnist, 10)
y_test_mnist = tf.keras.utils.to_categorical(y_test_mnist, 10)
```

```python
# Preprocess CIFAR-10 data
x_train_cifar = x_train_cifar / 255.0
x_test_cifar = x_test_cifar / 255.0
y_train_cifar = tf.keras.utils.to_categorical(y_train_cifar, 10)
y_test_cifar = tf.keras.utils.to_categorical(y_test_cifar, 10)
```

```python
# Define a function to create CNN models
def create_cnn(input_shape, activation_fn):
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation=activation_fn, input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation=activation_fn),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation=activation_fn),
        layers.Flatten(),
        layers.Dense(64, activation=activation_fn),
        layers.Dense(10, activation='softmax')
    ])
    return model
```

```python
# Define activation functions
activation_functions = ['sigmoid', 'tanh', 'relu', 'elu', 'selu']
```

```python
# Training parameters
batch_size = 64
epochs = 10
```

```python
# Results storage
results = {}
for dataset_name, (x_train, y_train, x_test, y_test, input_shape) in {
    "MNIST": (x_train_mnist, y_train_mnist, x_test_mnist, y_test_mnist, x_train_mnist.shape[1:]),
    "CIFAR10": (x_train_cifar, y_train_cifar, x_test_cifar, y_test_cifar, x_train_cifar.shape[1:]),
}.items():
    print(f"\n\nTraining on {dataset_name} dataset\n")
    results[dataset_name] = {}
    for activation_fn in activation_functions:
        print(f"\nUsing activation function: {activation_fn}\n")

        # Create and compile model
        model = create_cnn(input_shape, activation_fn)
        model.compile(optimizer='adam',
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])

        # Train model
        history = model.fit(
            x_train, y_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_data=(x_test, y_test),
            verbose=1
        )

        # Save results
        results[dataset_name][activation_fn] = history.history

        # Plot training and validation loss and accuracy
        plt.figure(figsize=(12, 4))

        # Loss plot
        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title(f'{dataset_name} - {activation_fn} - Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()

        # Accuracy plot
        plt.subplot(1, 2, 2)
        plt.plot(history.history['accuracy'], label='Training Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title(f'{dataset_name} - {activation_fn} - Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()

        plt.show()
```

```python
# Analyze results
for dataset_name, dataset_results in results.items():
    print(f"\n\nBest activation function for {dataset_name}:\n")
    best_activation = max(dataset_results.keys(), key=lambda fn: max(dataset_results[fn]['val_accuracy']))
    best_val_accuracy = max(dataset_results[best_activation]['val_accuracy'])
    print(f"{best_activation} with validation accuracy: {best_val_accuracy:.4f}")
```
